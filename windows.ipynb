```python
import pandas as pd
```

```python
location = 'data/'
events1 = location + 'eventsFeatures1.csv'
events2 = location + 'eventsFeatures2.csv'
events3 = location + 'eventsFeatures3.csv'
events4 = location + 'eventsFeatures4.csv'
installs1 = location + 'installsFeatures1.csv'
installs2 = location + 'installsFeatures2.csv'
installs3 = location + 'installsFeatures3.csv'
installs4 = location + 'installsFeatures4.csv'
clicks1 = location + 'clicksFeatures1.csv'
clicks2 = location + 'clicksFeatures2.csv'
clicks3 = location + 'clicksFeatures3.csv'
clicks4 = location + 'clicksFeatures4.csv'
auctions1 = location + 'auctionsFeatures1.csv'
auctions2 = location + 'auctionsFeatures2.csv'
auctions3 = location + 'auctionsFeatures3.csv'
auctions4 = location + 'auctionsFeatures4.csv'
```

```python
dtypeEvents = {
    'device_id': 'str',
    'eventsCount': 'int64',
    'eventsMostFreqDay': 'int64',
    'eventsMeanInterval': 'float64',
    'eventsAproxFreq': 'float64',
    'eventsStdDevInterval': 'float64',
    'eventsLast': 'float64'
}
dtypeInstalls = {
    'device_id': 'str',
    'installsCount': 'int64',
    'installsMostFreqDay': 'int64',
    'installsMeanInterval': 'float64',
    'installsAproxFreq': 'float64',
    'installsStdDevInterval': 'float64',
    'installsLast': 'float64'
}
dtypeClicks = {
    'device_id': 'str',
    'clicksCount': 'int64',
    'clicksMostFreqDay': 'int64',
    'clicksMeanInterval': 'float64',
    'clicksAproxFreq': 'float64',
    'clicksStdDevInterval': 'float64',
    'clicksLast': 'float64'
}
dtypeAuctions = {
    'device_id': 'str',
    'auctionsCount': 'int64',
    'auctionsMostFreqDay': 'int64',
    'auctionsMeanInterval': 'float64',
    'auctionsAproxFreq': 'float64',
    'auctionsStdDevInterval': 'float64',
    'auctionsLast': 'float64'
}
```

Leo los Features de Events de cada ventana por separado

```python
eventsFeatures1 = pd.read_csv(events1, usecols=dtypeEvents.keys(), dtype=dtypeEvents)
eventsFeatures2 = pd.read_csv(events2, usecols=dtypeEvents.keys(), dtype=dtypeEvents)
eventsFeatures3 = pd.read_csv(events3, usecols=dtypeEvents.keys(), dtype=dtypeEvents)
eventsFeatures4 = pd.read_csv(events4, usecols=dtypeEvents.keys(), dtype=dtypeEvents)
installsFeatures1 = pd.read_csv(installs1, usecols=dtypeInstalls.keys(), dtype=dtypeInstalls)
installsFeatures2 = pd.read_csv(installs2, usecols=dtypeInstalls.keys(), dtype=dtypeInstalls)
installsFeatures3 = pd.read_csv(installs3, usecols=dtypeInstalls.keys(), dtype=dtypeInstalls)
installsFeatures4 = pd.read_csv(installs4, usecols=dtypeInstalls.keys(), dtype=dtypeInstalls)
clicksFeatures1 = pd.read_csv(clicks1, usecols=dtypeClicks.keys(), dtype=dtypeClicks)
clicksFeatures2 = pd.read_csv(clicks2, usecols=dtypeClicks.keys(), dtype=dtypeClicks)
clicksFeatures3 = pd.read_csv(clicks3, usecols=dtypeClicks.keys(), dtype=dtypeClicks)
clicksFeatures4 = pd.read_csv(clicks4, usecols=dtypeClicks.keys(), dtype=dtypeClicks)
auctionsFeatures1 = pd.read_csv(auctions1, usecols=dtypeAuctions.keys(), dtype=dtypeAuctions)
auctionsFeatures2 = pd.read_csv(auctions2, usecols=dtypeAuctions.keys(), dtype=dtypeAuctions)
auctionsFeatures3 = pd.read_csv(auctions3, usecols=dtypeAuctions.keys(), dtype=dtypeAuctions)
auctionsFeatures4 = pd.read_csv(auctions4, usecols=dtypeAuctions.keys(), dtype=dtypeAuctions)
```

Agrego un sufijo según la ventana a la que representa

```python
eventsFeatures1['device_id'] = eventsFeatures1['device_id'] + '_1'
eventsFeatures2['device_id'] = eventsFeatures2['device_id'] + '_2'
eventsFeatures3['device_id'] = eventsFeatures3['device_id'] + '_3'
eventsFeatures4['device_id'] = eventsFeatures4['device_id'] + '_4'
installsFeatures1['device_id'] = installsFeatures1['device_id'] + '_1'
installsFeatures2['device_id'] = installsFeatures2['device_id'] + '_2'
installsFeatures3['device_id'] = installsFeatures3['device_id'] + '_3'
installsFeatures4['device_id'] = installsFeatures4['device_id'] + '_4'
clicksFeatures1['device_id'] = clicksFeatures1['device_id'] + '_1'
clicksFeatures2['device_id'] = clicksFeatures2['device_id'] + '_2'
clicksFeatures3['device_id'] = clicksFeatures3['device_id'] + '_3'
clicksFeatures4['device_id'] = clicksFeatures4['device_id'] + '_4'
auctionsFeatures1['device_id'] = auctionsFeatures1['device_id'] + '_1'
auctionsFeatures2['device_id'] = auctionsFeatures2['device_id'] + '_2'
auctionsFeatures3['device_id'] = auctionsFeatures3['device_id'] + '_3'
auctionsFeatures4['device_id'] = auctionsFeatures4['device_id'] + '_4'
```

Concateno todas las ventanas en un mismo DataFrame

```python
eventsFeatures = pd.concat([eventsFeatures1, eventsFeatures2, eventsFeatures3, eventsFeatures4])
eventsFeatures1 = 0
eventsFeatures2 = 0
eventsFeatures3 = 0
eventsFeatures4 = 0
installsFeatures = pd.concat([installsFeatures1, installsFeatures2, installsFeatures3, installsFeatures4])
installsFeatures1 = 0
installsFeatures2 = 0
installsFeatures3 = 0
installsFeatures4 = 0
clicksFeatures = pd.concat([clicksFeatures1, clicksFeatures2, clicksFeatures3, clicksFeatures4])
clicksFeatures1 = 0
clicksFeatures2 = 0
clicksFeatures3 = 0
clicksFeatures4 = 0
auctionsFeatures = pd.concat([auctionsFeatures1, auctionsFeatures2, auctionsFeatures3, auctionsFeatures4])
auctionsFeatures1 = 0
auctionsFeatures2 = 0
auctionsFeatures3 = 0
auctionsFeatures4 = 0
```

Junto todas las features con un grandioso outer join.

```python
features = pd.merge(eventsFeatures, installsFeatures, how='outer')
eventsFeatures = 0
installsFeatures = 0
features = pd.merge(features, clicksFeatures, how='outer')
clicksFeatures = 0
features = pd.merge(features, auctionsFeatures, how='outer')
auctionsFeatures = 0
```

```python
features = features.reset_index(drop=True)
```

Guardo el nuevo DataFrame en un csv.

```python
features.to_csv('data/features.csv', index=False)
```
