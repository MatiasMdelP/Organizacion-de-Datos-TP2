```python
import pandas as pd

firstDay = pd.Timestamp(2019, 4, 18)
lastDay = pd.Timestamp(2019, 4, 27)

location = 'data/'
installsFile = location + 'installs.csv'
eventsFile = location + 'events.csv'
clicksFile = location + 'clicks.csv'
auctionsFile = location + 'auctions.csv'
```

```python
dtype = {
    'advertiser_id': 'category',
    'source_id': 'category',
    'created': 'str',
    'latitude': 'float32',
    'longitude': 'float32',
    'wifi_connection': 'bool',
    'carrier_id': 'category',
    'os_minor': 'category',
    'os_major': 'category',
    'agent_device': 'category',
    'specs_brand': 'category',
    'brand': 'category',
    'timeToClick': 'float32',
    'touchX': 'str',
    'touchY': 'str',
    'ref_type': 'category',
    'ref_hash': 'int64',
}
```

```python
clicks=pd.read_csv(clicksFile, usecols=dtype.keys(), dtype=dtype, parse_dates=['created'])
```

```python
clicks.info()
```

```python
clicks=clicks.drop(['latitude','longitude','agent_device','touchX','touchY','brand','carrier_id','os_minor','os_major'],axis=1)
```

```python
onehot=pd.get_dummies(clicks['advertiser_id'])
onehot.columns=['clicks_0_advertiser','clicks_1_advertiser','clicks_2_advertiser','clicks_3_advertiser','clicks_4_advertiser']
clicks=pd.concat([clicks,onehot],axis=1)
onehot=pd.get_dummies(clicks['source_id'])
onehot.columns=['clicks_0_source_id','clicks_1_source_id','clicks_2_source_id','clicks_3_source_id','clicks_4_source_id','clicks_5_source_id','clicks_6_source_id','clicks_7_source_id','clicks_8_source_id']
clicks=pd.concat([clicks,onehot],axis=1)
onehot=pd.get_dummies(clicks['specs_brand'])
onehot.columns=['clicks_0_specs','clicks_1_specs','clicks_2_specs','clicks_3_specs','clicks_4_specs']
clicks=pd.concat([clicks,onehot],axis=1)
onehot=pd.get_dummies(clicks['ref_type'])
onehot.columns=['clicks_0_ref_type','clicks_1_ref_type']
clicks=pd.concat([clicks,onehot],axis=1)
onehot=pd.get_dummies(clicks['wifi_connection'])
onehot.columns=['clicks_wifi_false','clicks_wifi_true']
clicks=pd.concat([clicks,onehot],axis=1)

clicks=clicks.drop(['ref_type','advertiser_id','source_id','created','specs_brand','timeToClick','wifi_connection'],axis=1)
```

```python
dtype = {
    'created': 'str',
    'application_id': 'category',
    'ref_type': 'category',
    'ref_hash': 'int64',
    'click_hash': 'str',
    'attributed': 'bool',
    'implicit': 'bool',
    'device_brand': 'float64',
    'device_model': 'float64',
    'session_user_agent': 'str',
    'user_agent': 'str',
    'event_uuid': 'str',
    'kind': 'str',
    'wifi': 'str',
    'trans_id': 'str',
    'ip_address': 'int64',
    'device_language': 'category',
}
```

```python
installs=pd.read_csv(installsFile, usecols=dtype.keys(), dtype=dtype, parse_dates=['created'])
```

```python
installs.info()
```

```python
installs=installs.drop(['application_id','ip_address','kind','device_brand','device_model','session_user_agent','event_uuid','trans_id','click_hash','user_agent','device_language','created','ref_type'],axis=1)
```

```python
installs['wifi']=installs['wifi'].fillna('unknown')
```

```python
onehot=pd.get_dummies(installs['attributed'])
onehot.columns=['installs_attributed_false','installs_attributed_true']
installs=pd.concat([installs,onehot],axis=1)
onehot=pd.get_dummies(installs['implicit'])
onehot.columns=['installs_implicit_false','installs_implicit_true']
installs=pd.concat([installs,onehot],axis=1)
onehot=pd.get_dummies(installs['wifi'])
onehot.columns=['installs_wifi_false','installs_wifi_true','installs_wifi_unknown']
installs=pd.concat([installs,onehot],axis=1)

installs=installs.drop(['attributed','implicit','wifi'],axis=1)
```

```python
dtype = {
    'date': 'str',
    'event_id': 'int8',
    'ref_type': 'category',
    'ref_hash': 'int64',
    'application_id': 'int64',
    'attributed': 'bool',
    'device_os_version': 'float64',
    'device_brand': 'float64',
    'device_model': 'float64',
    'device_city': 'float64',
    'session_user_agent': 'float64',
    'trans_id': 'str',
    'user_agent': 'float64',
    'event_uuid': 'str',
    'carrier': 'float64',
    'kind': 'float64',
    'device_os': 'category',
    'wifi': 'bool',
    'connection_type': 'category',
    'ip_address': 'int64',
    'device_language': 'category'
}
```

```python
events=pd.read_csv(eventsFile, usecols=dtype.keys(), dtype=dtype, parse_dates=['date'])
```

```python
events.info()
```

```python
events=events.astype({'device_os':'object','connection_type':'object'})
```

```python
events['device_os']=events['device_os'].fillna('unknown')
events['connection_type']=events['connection_type'].fillna('unknown')
```

```python
events=events.drop(['date','event_id','device_os_version','device_brand','device_model','device_city','session_user_agent','trans_id','user_agent','event_uuid','carrier','kind','ip_address','device_language','application_id'],axis=1)
```

```python
onehot=pd.get_dummies(events['attributed'])
onehot.columns=['event_attributed_false','event_attributed_true']
events=pd.concat([events,onehot],axis=1)
onehot=pd.get_dummies(events['ref_type'])
onehot.columns=['event_ref_type0','event_ref_type1']
events=pd.concat([events,onehot],axis=1)
onehot=pd.get_dummies(events['device_os'])
onehot.columns=['event_device_os0','event_device_os0','unknown']
events=pd.concat([events,onehot],axis=1)
onehot=pd.get_dummies(events['wifi'])
onehot.columns=['event_wifiFalse','event_wifiTrue']
events=pd.concat([events,onehot],axis=1)
onehot=pd.get_dummies(events['connection_type'])
onehot.columns=['event_connection_type_Cable/DSL','event_connection_type_Cellular','event_connection_type_Corporate','event_connection_type_Dialup','event_connection_type_Unknown']
events=pd.concat([events,onehot],axis=1)

events=events.drop(['ref_type','attributed','device_os','wifi','connection_type'],axis=1)
```

```python
dtype = {
    'device_id': 'int64',
    'date': 'str',
    'ref_type_id': 'category',
    'source_id': 'category',
}
```

```python
auctions=pd.read_csv(auctionsFile, usecols=dtype.keys(), dtype=dtype, parse_dates=['date'])
```

```python
auctions.info()
```

```python
onehot=pd.get_dummies(auctions['ref_type_id'])
onehot.columns=['auctions_ref_type_id1','auctions_ref_type_id7']
auctions=pd.concat([auctions,onehot],axis=1)
onehot=pd.get_dummies(auctions['source_id'])
onehot.columns=['auctions_source_id0','auctions_source_id1','auctions_source_id2','auctions_source_id3','auctions_source_id4','auctions_source_id5','auctions_source_id6','auctions_source_id7','auctions_source_id8','auctions_source_id9']
auctions=pd.concat([auctions,onehot],axis=1)



auctions=auctions.drop(['ref_type_id','source_id','date'],axis=1)
```

```python
clicks=clicks.rename(columns={'ref_hash':'device_id'})
installs=installs.rename(columns={'ref_hash':'device_id'})
events=events.rename(columns={'ref_hash':'device_id'})
```
